```python
record = " EmployeeName :  john doe , Email : JOHN.DOE@company.COM , ID :  emp_0087 "
record_clean=record.strip()

print(record_clean.index('john doe'))
print(record_clean.index('JOHN.DOE@company.COM'))
print(record_clean.index('emp_0087'))
print(record_clean.index('COM'))

print('Name:',record_clean[16:24].title() )
print('Email:',record_clean[35:55].lower())
print('Emp_ID:',record_clean[-4:])

print('Name:',record_clean[16:24].title(),'\nEmail:',record_clean[35:55].lower(),'\nEmp_ID:',record_clean[-4:])
```
