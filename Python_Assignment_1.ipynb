# 2. String Slicing and Indexing: 


```python
# a. Print the first character of the string. 
First_character=out_put_string[0]

print('First character of the string: ',First_character)

# b. Print the last character of the string. 

Last_character=out_put_string[-1]

print('Last character of the string: ',Last_character)

# c. Print the first 5 characters of the string. 

First_5_character=out_put_string[:6]

print('First 5 character of the string: ',First_5_character)

# d. Print the last 11 characters of the string. 


Last_11_character=out_put_string[-11:]


# e. Print the string in reverse. 

Reverse_string=out_put_string[::-1]

print('String in reverse: ',Reverse_string)

# f. Use slicing and print the word “Python” from the existing string.

print('Python Index Postion: ',out_put_string.index('Python'))

print('Word Python from string : ',out_put_string[25:31])

```

# 3. String Methods: 
Write a Python program that takes a string, strM = “Python beginner tutorial” and 
perform the following tasks: 

```python
strM='Python beginner tutorial'
#  a. Convert the sentence to uppercase. 

uppercase= strM.upper()
print('Uppercase sentence: ',uppercase)

#  b. Convert the sentence to lowercase. 

lowercase=uppercase.lower()
print('Lowercase sentence: ',lowercase)

#c. Use Capitalize and return the sentence to the original input form.

Original_Format=lowercase.capitalize()

print('Capitalize and return the sentence to the original input form: ',Original_Format)

#d. Count the total number of occurrences of character ‘t’  in the string.

Total_No_Of_Char_Occurences=Original_Format.count('t')

print('Total number of occurrences of character ‘t’: ',Total_No_Of_Char_Occurences)

# e. Replace all occurrences of “Python” with “Machine Learning”  in the input string strM = “Python beginner tutorial”
Replace=strM.replace('Python',' Machine Learning')

print( 'Replace all occurrences of “Python” with “Machine Learning”: ',Replace)
```
